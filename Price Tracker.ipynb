```python
from bs4 import BeautifulSoup
from selenium import webdriver
name = input("Enter the product you want to search: ")
def flipkart(name):
    global flipkart
    fkurl = f'https://www.flipkart.com/search?q={name}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
    driver=webdriver.Chrome()
    driver.get(fkurl)
    fksoup = BeautifulSoup(driver.page_source,'html.parser')
    driver.quit()
    if (fksoup.find_all('div',class_='KzDlHZ')):
        fk_page = fksoup.find_all('div',class_='KzDlHZ')
        fk_pagelength = int(len(fk_page))
        for i in range (0,fk_pagelength):
            fk_name = fksoup.find_all('div',class_='KzDlHZ')[i].text.strip().upper()
            if name.upper() in fk_name:
                fk_price = fksoup.find_all('div', class_='Nx9bqj _4b5DiR')[i].text.strip()
                fk_rating = fksoup.find_all('div', class_= 'XQDdHH')[i].text.strip()
                fk_href = fksoup.find_all('a', class_ = 'CGtC98')[i].get('href').strip()
                fk_link = "https://flipkart.com" + fk_href
                print("Searching in Flipkart.......",end = '\n')
                print("> Product Found!",end = '\n')
                print( "> Product Name: ",fk_name, end = '\n')
                print("> Product Price: ",fk_price, end = '\n')
                print("> Product Rating: ",fk_rating," out of 5 stars",end = '\n\n')
                break
            else:
                print("Product not found in Flipkart")
                fk_link = None
                fk_price = 0
    elif (fksoup.find_all('a',class_='wjcEIp')):
        fk_page = fksoup.find_all('a',class_='wjcEIp')
        fk_pagelength = int(len(fk_page))
        for i in range(0,fk_pagelength):                   
            fk_name = fksoup.find_all('a',class_='wjcEIp')[i].text.strip().upper()
            if name.upper() in fk_name:
                fk_price = fksoup.find_all('div', class_='Nx9bqj')[i].text.strip()
                fk_rating = fksoup.find_all('div', class_= 'XQDdHH')[i].text.strip()
                fk_href = fksoup.find_all('a', class_= 'wjcEIp')[i].get('href').strip()
                print("Searching in Flipkart.......",end = '\n')
                print("> Product Found!",end = '\n')
                print( "> Product Name: ",fk_name, end = '\n')
                print("> Product Price: ",fk_price, end = '\n')
                print("> Product Rating: ",fk_rating," out of 5 stars",end = '\n\n')
                break
            else:
                print("Product not found in Flipkart")
                fk_link = None
                fk_price = 0 
    return fk_price,fk_link
def amazon(name):
    global amazon
    name1 = name.replace(" " , "-")
    name2 = name.replace(" ", "+")
    amzurl = f'https://www.amazon.in/{name1}/s?k={name2}'
    driver = webdriver.Chrome()
    driver.get(amzurl)
    amzsoup = BeautifulSoup(driver.page_source,'html.parser')
    driver.quit()
    amz_page = amzsoup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    amz_pagelength= int(len(amz_page))
    for i in range (0,amz_pagelength):
        name = name.upper()
        amz_name = amzsoup.find_all('span',class_='a-size-medium a-color-base a-text-normal')[i].text.strip().upper()
        if name in amz_name :
            amz_price = amzsoup.find_all('span', class_='a-price-whole')[i].text
            amz_rating = amzsoup.find_all('span', class_= 'a-icon-alt')[i].text
            amz_href = amzsoup.find_all('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal')[i].get('href')
            amz_link = "https://amazon.in" + amz_href
            break
        else:
            i+=1
            i=int(i)
            if i == amz_pagelength:
                amz_price = 0
                print("Product not found in Amazon")     
    print("Searching in Amazon.......",end = '\n')
    print("> Product Found!",end = '\n')
    print( "> Product name: ",amz_name, end = '\n')
    print("> Product Price:₹",amz_price, end = '\n')
    print("> Product Rating: ",amz_rating,end = '\n')
    return amz_price,amz_link
flk_price,flk_link = flipkart(name)
amzn_price,amzn_link = amazon(name)
def convert(a):
    b=a.replace(" ",'')
    c=b.replace("INR",'')
    d=c.replace(",",'')
    f=d.replace("₹",'')
    g=int(float(f))
    return g
flk_price = convert(flk_price)
amzn_price = convert(amzn_price)
print('\n\n')
if(flk_price > amzn_price):
    print("The product is cheaper in Amazon.",end = '\n')
    print("Product Link: ", amzn_link)
elif(flk_price < amzn_price):
    print("The product is cheaper in Flipkart.",end = '\n')
    print("Product Link: ",flk_link)
elif(flk_price == amzn_price):
    print("The price is same in both the websites")
    print('\n')
    print("Flipkart link: ",flk_link)
    print("Amazon Link: ",amzn_link)
    
```
